## Lectura secuencial de la secuencia, sin cargar todo de una sola vez para evitar utilar toda la RAM

```python

recs = SeqIO.parse(gzip.open('SRR003265.filt.fastq.gz', 'rt', encoding='utf-8'), 'fastq')
count = defaultdict(int) 
# simil a un diccionario vacio donde las claves se van creando a medida que se le asigna,
#y el valor por default a una clave que sea creada pero a la cual no se le asigna un valor es 0

for rec in recs:
    for letter in rec.seq:
        count[letter] += 1
tot = sum(count.values())
for letter, count in count.items():
    print('%s: %.2f %d' % (letter, 100. * count / tot, count))
```

```python
#inicializo un diccionario que contenga los datos
recs = SeqIO.parse(gzip.open('SRR003265.filt.fastq.gz', 'rt', encoding='utf-8'), 'fastq')
n_conteo = defaultdict(int)

for rec in recs:
    for i, letra in enumerate(rec.seq):
        posicion = i + 1 #El conteo arranca en 0
        if letra =="N":
            n_conteo[posicion] += 1 
            
            

seq_len = max(n_conteo.keys())
positions = range(1, seq_len + 1)
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(positions, [n_conteo[x] for x in positions])
ax.set_xlim(1, seq_len)
pass
```

## Calidad de las lecturas

```python
recs = SeqIO.parse(gzip.open('SRR003265.filt.fastq.gz', 'rt', encoding='utf-8'), 'fastq')
calidad_lecturas_posicion = defaultdict(list)

for rec in recs:
    for i, calidad in enumerate(rec.letter_annotations['phred_quality']):
        if i < 25 or calidad == 40:
            continue
        posicion = i +1 
        calidad_lecturas_posicion[posicion].append(calidad)
        
        
vps = []
poses = list(calidad_lecturas_posicion.keys())
#poses.sort()
#for pos in poses:
 #   vps.append(calidad_lecturas_posicion[posicion])
    
    
    
    
```

```python
vps = []
posiciones = list(calidad_lecturas_posicion.keys())
posiciones.sort()

for posicion in posiciones:
    vps.append(calidad_lecturas_posicion[posicion])
    
fig, ax = plt.subplots(figsize=(16,9))
sns.boxplot(data=vps, ax=ax)
ax.set_xticklabels([str(x) for x in range(26, max(calidad_lecturas_posicion.keys()) + 1)])
pass
```
