# Blast 

```python
#Biopython version para visualizar 1.79 // en 1.8 no funciona
from Bio import SeqIO
proteins_sequence = SeqIO.read("protein_seq.fasta","fasta")
len(proteins_sequence.seq)
proteins_sequence.seq
```

```python
from Bio.Blast import NCBIWWW
resultados = NCBIWWW.qblast("blastp","pdb",proteins_sequence.seq)
```

```python
from Bio import SearchIO
blast_records = SearchIO.read(resultados, 'blast-xml')
print(blast_records[0:10])
```

```python
for blast_record in blast_records:
    print(f"Sequence ID: {blast_record.id}")
    print(f"description: {blast_record.description}")
    print(f"E value: {blast_record[0].evalue}")
    print(f"Bit Score:  {blast_record[0].bitscore}")
    print(f"alignment:\n{blast_record[0].aln}")
    print()
```
